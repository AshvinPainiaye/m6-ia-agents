# Mini-atelier transverse  
## Programmation orientée objectifs et boucles décisionnelles IA  
**Rattaché au module 6 – Amélioration continue d’une solution IA**

## 1. Contexte professionnel

En tant que développeur·se IA, vous intervenez sur une solution déjà déployée intégrant un modèle d’intelligence artificielle.  
Le modèle produit des réponses fonctionnelles, mais les retours utilisateurs montrent :

- des réponses parfois imprécises,
- des cas où le modèle ne devrait pas répondre,
- des situations nécessitant une reformulation ou une escalade humaine.

Votre rôle n’est pas de réentraîner le modèle, mais de **concevoir une logique applicative** capable de décider automatiquement **quoi faire** à partir de la sortie du modèle.

## 2. Objectif du mini-atelier

Mettre en place une **boucle décisionnelle** autour d’un modèle IA existant afin de :

- contrôler la qualité des réponses,
- introduire des règles de décision explicites,
- renforcer la robustesse et la responsabilité du système.

## 3. Compétences mobilisées

- Orchestration d’un composant IA dans une application.
- Gestion d’un état applicatif minimal.
- Implémentation de règles de décision conditionnelles.
- Analyse critique des limites d’un modèle IA.
- Documentation de choix techniques.

## 4. Modalités pédagogiques

- Travail individuel ou en binôme.
- Durée indicative : **6 à 8 heures**.
- Atelier transverse, hors évaluation RNCP.
- Restitution écrite et démonstration fonctionnelle.

## 5. Livrables attendus

1. Un script Python implémentant la boucle décisionnelle.
2. Une documentation courte expliquant :
   - les règles de décision,
   - les seuils utilisés,
   - les limites du système.
3. Un schéma simple du fonctionnement global.
